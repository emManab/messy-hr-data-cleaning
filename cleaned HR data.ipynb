#### word2number library for AGE & Salary

```python
from word2number import w2n
def convert_to_int(x):
    try:
        return int(x)  # if it's already a numeric string
    except ValueError:
        try:
            return w2n.word_to_num(x)  # if it's written in words
        except:
            return None  # fallback if conversion fails
data['Age'] = data['Age'].apply(convert_to_int)
data.tail()

#ORR
# data['Age'] = data['age'].replace({'thirty': 30}) best for this data cz here only 'thirty' is there in it
```

```python
data['Salary'] = data['Salary'].apply(convert_to_int)
data.tail()
#ORR
# data['Salary'] = data['Salary'].replace({'SIXTY THOUSAND': 60000}) suitable for this data
```

#### Changing Datatype of AGE & Salary to INT64

```python
data['Age'] = data['Age'].astype('Int64')
data["Salary"] = data['Salary'].astype('Int64')
print(data.dtypes)
```

#### Changing Format od Phone Number

```python
data['Phone Number'] = data['Phone Number'].str.replace('-' , '')
data['Phone Number']
```

#### We dont need any mathematical operation on Phone NO. so object is fine

```python
data.head()
```

```python
data['Joining Date'] = data['Joining Date'].apply(
    lambda x: parser.parse(x) if isinstance(x, str) else x
)
data.head()
data.dtypes
```

#### Changing Performance Score data type to Categorical Data

```python
data['Performance Score'] = data['Performance Score'].astype('category')
# say A is best, F is worst:
data['Performance Score'] = pd.Categorical(
    data['Performance Score'],
    categories=['A', 'B', 'C','E','F'],
    ordered=True
)
data.dtypes
```

#### Now you can Filter Data in order A > B > C > D and F is fail / poor

```python
# Using bitwise OR
data[(data['Performance Score'] == "A") | (data['Performance Score'] == "B")]
```

```python
data[data['Performance Score'].isin(["C", "F"])]
```
